```python
import pandas as pd
```

```python
data = pd.read_csv("vine_table.csv")
```

```python
df = pd.DataFrame(data)
```

```python
# vine_table dataframe
df
```

```python
# data filtered for 20 or more total votes
total_votes_df = df[df["total_votes"] >= 20.0]
```

```python
total_votes_df
```

```python
# creating a helpful_percentage column
df["helpful_percentage"] = (df["helpful_votes"] / df["total_votes"]) * 100
```

```python
df
```

```python
# percentage of helpful_votes is equal or greater than 50%
helpful_vote_percentage_df = df[df["helpful_percentage"] >= 50.0]
```

```python
helpful_vote_percentage_df
```

```python
# data filtered for paid vine reviews
vine_paid_df = helpful_vote_percentage_df[helpful_vote_percentage_df["vine"] == "Y"]
```

```python
# data filtered for unpaid vine reviews
vine_unpaid_df = helpful_vote_percentage_df[helpful_vote_percentage_df["vine"] == "N"]
```

```python
vine_paid_df
```

```python
vine_unpaid_df
```

```python
# total number of vine paid reviews
vine_paid_df["review_id"].count()
```

```python
# total number of vine unpaid reviews
vine_unpaid_df["review_id"].count()
```

```python
vine_paid_fivestar_df = vine_paid_df[vine_paid_df["star_rating"] >= 5.0]
```

```python
vine_unpaid_fivestar_df = vine_unpaid_df[vine_unpaid_df["star_rating"] >= 5.0]
```

```python
# total number of vine paid fivestar reviews
vine_paid_fivestar_df["review_id"].count()
```

```python
# total number of vine unpaid fivestar reviews
vine_unpaid_fivestar_df["review_id"].count()
```

```python
vine_paid_fivestar_percentage = (vine_paid_fivestar_df["review_id"].count() / vine_paid_df["review_id"].count()) * 100
```

```python
# the percentage of paid fivestar reviews
vine_paid_fivestar_percentage
```

```python
vine_unpaid_fivestar_percentage = (vine_unpaid_fivestar_df["review_id"].count() / vine_unpaid_df["review_id"].count()) * 100
```

```python
# percentage of unpaid fivestar reviews
vine_unpaid_fivestar_percentage
```
